```python
from typing import List, Tuple

#Create a new Map
my_map = dict()

#1.Adding key-value pairs
print("===BASIC OPERATIONS===")

# Adding key-value pairs
my_map["name"] = "Jonh Doe"
my_map["age"] = "30"
my_map["city"] = "New York"
my_map["occupation"] = "Software Engineer"

print(f"Initial Map:{my_map}")
print(f"Size:{len(my_map)}")

#Accessing values
print(f"Name:{my_map['name']}")
print(f"Age:{my_map['age']}")

#updating values
my_map["age"] = 31
print(f"Updated age: {my_map['age']}")

#Checking existence
print(f"Has 'City:' {'City' in my_map}")
print(f"Has 'Country: {'Country' in my_map}")

```
